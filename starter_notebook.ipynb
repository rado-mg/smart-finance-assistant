### Foundation Data Processing Functions

````python
import pandas as pd

# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant

    🤖 AI Collaboration Prompt:
    "Create a function to load CSV transaction data with Date, Amount, Category,
    Description columns. Handle dollar signs in Amount, missing values, and
    data validation. Include clear business-focused error messages."

    Args:
        file_path (str): Path to CSV file or file object
    Returns:
        pandas.DataFrame: Cleaned transaction data
    """
    try:
        # Step 1: Load CSV file
        df = pd.read_csv(file_path)

        # Step 2: Validate required columns
        required_columns = {"Date", "Amount", "Category", "Description"}
        missing_columns = required_columns - set(df.columns)
        if missing_columns:
            raise ValueError(f"Missing required columns: {', '.join(missing_columns)}")

        # Step 3: Clean 'Amount' column
        # - Remove dollar signs and commas
        # - Convert to numeric (float), coercing invalid entries to NaN
        df['Amount'] = (
            df['Amount']
            .astype(str)  # Ensure values are strings
            .str.replace(r'[\$,]', '', regex=True)  # Remove $ and commas
        )
        df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

        # Step 4: Handle missing values
        df['Category'] = df['Category'].fillna("Unknown")
        df['Description'] = df['Description'].fillna("No description")

        # Step 5: Convert 'Date' column to datetime format
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

        # Drop rows where Date conversion failed or Amount is missing
        df = df.dropna(subset=['Date', 'Amount'])

        return df

    except FileNotFoundError:
        raise Exception(f"❌ Business Error: File not found at {file_path}. Please check the path.")
    except pd.errors.EmptyDataError:
        raise Exception("❌ Business Error: The CSV file is empty or invalid.")
    except Exception as e:
        raise Exception(f"❌ Business Error: Unexpected issue occurred - {str(e)}")
# Sample test with your dataset
try:
    clean_data = load_and_clean_transaction_data("sample_transactions.csv")
    print("✅ Cleaned dataset preview:")
    print(clean_data.head())
except Exception as e:
    print(e)
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights

    🤖 AI Collaboration Prompt:
    "Create a function that analyzes spending by category, calculates percentages,
    identifies top spending areas, and generates actionable financial insights
    formatted for business presentation."

    Args:
        df (pandas.DataFrame): Cleaned transaction DataFrame
                              Must contain Date, Amount, Category, Description
    Returns:
        dict: Analysis results and insights
    """
    try:
        # Step 1: Calculate total spending (net of refunds)
        total_spending = df['Amount'].sum()

        # Step 2: Refunds (negative amounts)
        refund_total = df[df['Amount'] < 0]['Amount'].abs().sum()

        # Step 3: Category breakdown
        category_totals = df.groupby('Category')['Amount'].sum().round(2)

        # Step 4: Percentages by category
        category_percentages = (category_totals / total_spending * 100).round(2)

        # Step 5: Identify top spending areas (Top 3 categories)
        top_categories = category_totals.sort_values(ascending=False).head(3).to_dict()

        # Step 6: Generate business insights
        insights = []

        if 'Groceries' in category_percentages and category_percentages['Groceries'] > 40:
            insights.append("Groceries represent a large share of spending. Consider meal planning or bulk buying.")

        if refund_total > 0:
            insights.append("Refunds detected. Track them separately to avoid confusion in monthly totals.")

        if total_spending > 3000:
            insights.append("Spending exceeds $3000. Review discretionary expenses to stay within budget.")

        if not insights:
            insights.append("Spending appears balanced across categories. Continue monitoring for unusual spikes.")

        # Step 7: Return structured analysis
        return {
            "Total Spending (Net of Refunds)": round(total_spending, 2),
            "Refund Impact": round(refund_total, 2),
            "Category Breakdown": category_totals.to_dict(),
            "Category Percentages (%)": category_percentages.to_dict(),
            "Top Spending Areas": top_categories,
            "Business Insights": insights
        }

    except Exception as e:
        return {"Error": f"❌ Business Error: Unable to analyze spending - {str(e)}"}
# Assuming you already cleaned the data with load_and_clean_transaction_data()
analysis = analyze_spending_patterns(clean_data)

print("📊 Spending Analysis Report")
print("---------------------------------")
print(f"Total Spending (Net of Refunds): ${analysis['Total Spending (Net of Refunds)']}")
print(f"Refund Impact: ${analysis['Refund Impact']}")

print("\nCategory Breakdown:")
for category, amount in analysis['Category Breakdown'].items():
    print(f" - {category}: ${amount}")

print("\nCategory Percentages:")
for category, pct in analysis['Category Percentages (%)'].items():
    print(f" - {category}: {pct}%")

print("\nTop Spending Areas:")
for category, amount in analysis['Top Spending Areas'].items():
    print(f" - {category}: ${amount}")

print("\nBusiness Insights:")
for tip in analysis['Business Insights']:
    print(f" - {tip}")
````

````python
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis

    🤖 AI Collaboration Prompt:
    "Based on spending analysis data, create professional financial
    recommendations. Include specific savings opportunities, spending
    pattern observations, and actionable advice formatted for a
    personal finance app user."

    Args:
        analysis_data (dict): Dictionary with spending analysis results
    Returns:
        str: Formatted recommendations report
    """
    try:
        # Extract key values
        total_spending = analysis_data.get("Total Spending (Net of Refunds)", 0)
        refund_impact = analysis_data.get("Refund Impact", 0)
        category_percentages = analysis_data.get("Category Percentages (%)", {})
        top_categories = analysis_data.get("Top Spending Areas", {})

        # Build recommendations
        recommendations = []
        recommendations.append("📊 Financial Recommendations Report")
        recommendations.append("-----------------------------------")

        # Spending overview
        recommendations.append(f"Total Spending (Net of Refunds): ${total_spending}")
        if refund_impact > 0:
            recommendations.append(f"Refunds reduced your net spending by ${refund_impact}.")

        # Top categories
        recommendations.append("\nTop Spending Areas:")
        for category, amount in top_categories.items():
            recommendations.append(f" - {category}: ${amount}")

        # Actionable advice rules
        if 'Groceries' in category_percentages and category_percentages['Groceries'] > 40:
            recommendations.append("\nAdvice: Groceries are a major expense. Consider meal planning, bulk buying, or discount retailers.")

        if total_spending > 3000:
            recommendations.append("Advice: Spending exceeds $3000. Review discretionary categories like entertainment or shopping.")

        if refund_impact > 0:
            recommendations.append("Advice: Track refunds separately to avoid confusion in monthly totals.")

        if not analysis_data.get("Business Insights"):
            recommendations.append("Advice: Spending appears balanced. Keep monitoring for unusual spikes.")

        # Join into a formatted string
        return "\n".join(recommendations)

    except Exception as e:
        return f"❌ Business Error: Unable to generate recommendations - {str(e)}"
# Assuming you already ran analyze_spending_patterns()
recommendations = generate_financial_recommendations(analysis)
print(recommendations)
````
