Week 9: Simplifying step 1

````python
import pandas as pd

# Step 1: Load CSV file safely
try:
    df = pd.read_csv("transactions.csv")
    print("✅ CSV file loaded successfully.\n")
except FileNotFoundError:
    raise SystemExit("❌ Error: CSV file not found. Please check the file path.")

# Step 2: Inspect raw dataset
print("📊 First 5 rows of raw data:\n", df.head(), "\n")

# Step 3: Clean 'Amount' column in one go
df['Amount'] = (
    pd.to_numeric(
        df['Amount'].astype(str).str.replace(r'[\$,]', '', regex=True),
        errors='coerce'
    )
)

# Step 4: Validate and drop missing critical fields
df = df.dropna(subset=['Date', 'Amount'])
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])

# Step 5: Fill missing descriptions
df['Description'] = df['Description'].fillna("No description")

# Step 6: Show dataset info and summaries
print("🔍 Missing values per column:\n", df.isnull().sum(), "\n")
print("📈 Dataset Info:\n")
df.info()
print("\n📊 Summary Statistics for Amount:\n", df['Amount'].describe(), "\n")
print("📂 Unique Categories:\n", df['Category'].unique(), "\n")

# Final cleaned dataset preview
print("✅ Cleaned dataset preview:\n", df.head())

# Add your own comments about what you learned
'''I learned how to simplify a long code. It is useful because when the steps are written more compactly ,
we do not have to scroll as much, thus , it feels less heavy.Even though it is shorter
, the steps are still obvious, I mean start from loading, then cleaning, cheking and summarizing.'''
````

## 📈 Step 2: Create Core Analysis Functions

````python
# AI-generated analysis functions
# (Your AI collaboration code goes here)
import pandas as pd

def total_spending_by_category(df):
    """
    Business Explanation:
    This function calculates how much money was spent in each category.
    It helps users understand where their money is going (e.g., Food, Rent, Transport).

    Parameters:
        df (DataFrame): Cleaned transaction dataset with 'Category' and 'Amount' columns.

    Returns:
        Series: Total spending per category.
    """
    try:
        return df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
    except KeyError:
        print("❌ Error: Dataset must contain 'Category' and 'Amount' columns.")
        return None


def top_5_spending_categories(df):
    """
    Business Explanation:
    This function identifies the top 5 categories where the user spends the most.
    It highlights major spending areas for budgeting and financial planning.

    Parameters:
        df (DataFrame): Cleaned transaction dataset.

    Returns:
        Series: Top 5 categories ranked by spending.
    """
    totals = total_spending_by_category(df)
    if totals is not None:
        return totals.head(5)
    return None


def average_spending_per_transaction(df):
    """
    Business Explanation:
    This function calculates the average amount spent per transaction.
    It gives users a sense of their typical spending behavior.

    Parameters:
        df (DataFrame): Cleaned transaction dataset.

    Returns:
        float: Average spending value.
    """
    try:
        return df['Amount'].mean()
    except KeyError:
        print("❌ Error: Dataset must contain 'Amount' column.")
        return None


def identify_unusual_transactions(df, high_threshold=1000):
    """
    Business Explanation:
    This function flags unusual transactions:
    - Very high spending (above a threshold, default $1000)
    - Negative amounts (possible refunds or errors)
    It helps users spot anomalies or potential mistakes.

    Parameters:
        df (DataFrame): Cleaned transaction dataset.
        high_threshold (float): Value above which a transaction is considered unusual.

    Returns:
        DataFrame: Transactions flagged as unusual.
    """
    try:
        unusual = df[(df['Amount'] > high_threshold) | (df['Amount'] < 0)]
        return unusual
    except KeyError:
        print("❌ Error: Dataset must contain 'Amount' column.")
        return None
````
